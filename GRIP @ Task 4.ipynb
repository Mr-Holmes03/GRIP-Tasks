```python
#changing names of columns for better understanding.
df.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'day', 'gname':'Group', 'country_txt':'Country',
                   'region_txt':'Region', 'provstate':'State', 'city':'City', 'latitude':'Latitude', 'longitude':'Longitude',
                  'summary':'Summary', 'attacktype1_txt':'Attacktype', 'targtype1_txt':'Targettype', 'weaptype1_txt':'Weapon',
                  'nkill':'Kill', 'nwound':'Wound'}, inplace= True)
```

```python
#creating a new dataset having only useful features. (dropping non-required columns.)
df= df[['Year','Month','day','Country','State','Region','City','Latitude','Longitude','Attacktype','Kill','Wound',
       'target1','Summary','Group','Targettype','Weapon','motive','ransompaid','ransomamt']]
```

### Visualising Insights

```python
# Plot-1
#Plot for visualization of highest number of terrorist attacks.

plt.figure(figsize=(16,8))
sns.countplot(x='Region', data= df, order = df['Region'].value_counts().index)
plt.xticks(rotation = 90, fontsize= 10)
plt.title('Most Targeted Region', fontsize= 16)
plt.ylabel('Number of Attacks', fontsize=14)
plt.xlabel('Region', fontsize=14)
plt.show()
```

Observations:-

* Middle East & North Africa, South Asia region has highest numbers of terrorist attacks.
* Austrailasia & Oceania, Central Asia are least affected regions from terrorist attacks.

```python
# Plot-2
#Plot for visualising Type of Attack.

plt.figure(figsize=(16,8))
sns.countplot(x='Attacktype', data=df, order=df['Attacktype'].value_counts().index)
plt.xticks(rotation= 90, fontsize= 11)
plt.title('Type of Attack', fontsize=16)
plt.ylabel('Number of Attacks', fontsize=14)
plt.xlabel('Attack type', fontsize=14)
plt.show()
```

Observations:-

* Most of the terror attacks are excuted by using Explosives(Bombing).
* Very less terror attacks are done by Hijacking.


```python
# Plot-3
#Plot for visuals of number of Attacks per year.

plt.figure(figsize=(18,10))
sns.countplot(x= 'Year', data= df) # order= df['Year'].value_counts().index)
plt.title('Attacks per Year',fontsize= 16)
plt.xticks(rotation= 90, fontsize=12)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Attacks', fontsize=14)
plt.show()
```

Observations:
* Terror attacks are gradually increasing with year, But from 1992 to 1998 attacks were lowered in number and started rising again from 1999 and 2004.

```python
# Plot-4
# Plot for visuals of 10 years with maximum casualties in attacks.

plt.figure(figsize=(18,9))
kills= df.groupby('Year')['Kill'].sum()
top_10= kills.iloc[0:10]
top_10.plot(kind= 'bar')
plt.title('Top 10 years with Casualties in attacks', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Attacks', fontsize=14)
plt.show()
```

Observations:
* Number of deaths are also increasing every year in terrorist attacks, execpt in 1977.

```python
# Plot-5
#Plot for visualising 10 most ransom paid countries.

plt.figure(figsize=(18,9))
kills= df.groupby('Country')['ransomamt'].sum()
top_10= kills.sort_values(ascending= False). iloc[0:10]
top_10.plot(kind = 'bar')
plt.title('Top 10 Ransom Given Countries', fontsize=16)
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Amount of ransom paid', fontsize= 14)
plt.show()
```

Observations:
* Italy and Syria has given the most ransome as compared to other nations.

```python
# Plot-6
#Plot for Visuals of top 10 affected countries due to terrorist attacks.

plt.figure(figsize=(20,8))
sns.barplot(x= df['Country'].value_counts()[:10].index,y= df['Country'].value_counts()[:10], data= df)
plt.title('Top 10 affected countries', fontsize= 16)
plt.xticks(fontsize=12)
plt.xlabel('Countries', fontsize= 14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

Observations:
* Iraq and Pakistan are the most affected countries because of terrorism.

```python
# Plot-7
#Plot for Visuals of top 10 affected cities due to terrorist attacks.

plt.figure(figsize=(20,10))
sns.barplot(x= df['City'].value_counts()[:10].index,y= df['City'].value_counts()[:10], data= df)
plt.title('Top 10 affected cities', fontsize= 16)
plt.xticks(fontsize=12)
plt.xlabel('Cities', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

Observations:
* First bar in chart is of miscellaneous small cities, hence Baghdad is the most affected city by terrorist's activites.

```python
# Plot-8
#Plot for visuals of top 15 active terrorist organisations.

plt.figure(figsize=(20,8))
sns.barplot(x= df['Group'].value_counts()[0:15].index, y=df['Group'].value_counts()[0:15], data=df)
plt.title('Terrorist Organisations', fontsize= 16)
plt.xlabel('Organisations', fontsize= 14)
plt.xticks(rotation= 90, fontsize=12)
plt.ylabel('Count', fontsize=14)
plt.yscale('log')
plt.show()
```

Observation:
* Maximum attacks were done by Unknown small groups, but in famous organisation, Taliban is most active. 

```python
# Plot-9
#Plot for visualising most targeted properties.

plt.figure(figsize= (20,8))
sns.barplot(x= df['Targettype'].value_counts()[0:10].index, y= df['Targettype'].value_counts()[0:10], data=df)
plt.title('Targeted Properties', fontsize= 16)
plt.xlabel('Propert Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation= 90, fontsize=12)
plt.show()
```

Observation:
* Private citizens & property is the main target of terrorist organisation.

```python
# Plot-10
#Plot for casulties caused by terrorist organisation.

df[['Group','Kill']].groupby(['Group']).sum().drop('Unknown').sort_values('Kill',
    ascending=False)[0:10].plot(kind='bar',figsize=(20,10))
plt.title("Casulty caused by organisation", fontsize=16)
plt.xlabel("Terrorist organisation", fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.show()
```

Observation:
* Islamic State of Iraq and the Levant(ISIL) is responsible for maximum casualties among all other organisations.
